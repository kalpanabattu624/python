```python
#### import psycopg2
from tabulate import tabulate


def connection_database():
    return psycopg2.connect(
        host="localhost",
        database="school_students",
        user="postgres",
        password="Kalpana@15"
    )



def donor_id_exists(donor_id):
    conn = connection_database()
    cursor = conn.cursor()
    cursor.execute("SELECT donor_id FROM donors1 WHERE donor_id = %s", (donor_id,))
    result = cursor.fetchone()
    cursor.close()
    conn.close()
    return result is not None


def display_donors_table():
    conn = connection_database()
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM donors1")
    rows = cursor.fetchall()

    if rows:
        headers = ["Donor ID", "Name", "Blood Group", "Age", "Gender", "Last Donation"]
        print(tabulate(rows, headers=headers, tablefmt="grid"))
    else:
        print("No donors found")

    cursor.close()
    conn.close()



def add_student():
    try:
        donor_id = int(input("Enter donor ID: "))
        if donor_id_exists(donor_id):
            print("Donor ID already exists")
            return

        conn = connection_database()
        cursor = conn.cursor()

        name = input("Enter name: ")
        blood_group = input("Enter blood group: ")
        age = int(input("Enter age: "))
        gender = input("Enter gender: ")
        last_donation = input("Enter last donation date (YYYY-MM-DD): ")

        cursor.execute(
            """INSERT INTO donors1
               (donor_id, name, blood_group, age, gender, last_donation)
               VALUES (%s, %s, %s, %s, %s, %s)""",
            (donor_id, name, blood_group, age, gender, last_donation)
        )

        conn.commit()
        print(" Donor added successfully")

    except Exception as e:
        print("Error:", e)

    finally:
        try:
            cursor.close()
            conn.close()
        except:
            pass


def view_students():
    display_donors_table()



def update_student():
    try:
        print("\n--- Donor List ---")
        display_donors_table()

        donor_id = int(input("\nEnter donor ID to update: "))
        if not donor_id_exists(donor_id):
            print(" Donor ID not found")
            return

        conn = connection_database()
        cursor = conn.cursor()

        print("\nWhat do you want to update?")
        print("1. Name")
        print("2. Blood Group")
        print("3. Age")
        print("4. Gender")
        print("5. Last Donation Date")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            value = input("Enter new name: ")
            cursor.execute("UPDATE donors1 SET name=%s WHERE donor_id=%s", (value, donor_id))

        elif choice == 2:
            value = input("Enter new blood group: ")
            cursor.execute("UPDATE donors1 SET blood_group=%s WHERE donor_id=%s", (value, donor_id))

        elif choice == 3:
            value = int(input("Enter new age: "))
            cursor.execute("UPDATE donors1 SET age=%s WHERE donor_id=%s", (value, donor_id))

        elif choice == 4:
            value = input("Enter new gender: ")
            cursor.execute("UPDATE donors1 SET gender=%s WHERE donor_id=%s", (value, donor_id))

        elif choice == 5:
            value = input("Enter new last donation date (YYYY-MM-DD): ")
            cursor.execute("UPDATE donors1 SET last_donation=%s WHERE donor_id=%s", (value, donor_id))

        else:
            print("Invalid update choice")
            return

        conn.commit()
        print("Donor updated successfully")

    except Exception as e:
        print("Error:", e)

    finally:
        try:
            cursor.close()
            conn.close()
        except:
            pass

def delete_student():
    try:
        print("\n--- Donor List ---")
        display_donors_table()

        print("\nDelete Options:")
        print("1. Delete by Donor ID")
        print("2. Delete by Blood Group")
        print("3. Delete All Donors")
        print("4. Cancel")

        choice = int(input("Enter your choice: "))

        conn = connection_database()
        cursor = conn.cursor()

        if choice == 1:
            donor_id = int(input("Enter donor ID: "))
            if not donor_id_exists(donor_id):
                print("Donor ID not found")
                return
            cursor.execute("DELETE FROM donors1 WHERE donor_id=%s", (donor_id,))
            conn.commit()
            print(" Donor deleted successfully")

        elif choice == 2:
            bg = input("Enter blood group: ")
            cursor.execute("SELECT * FROM donors1 WHERE blood_group=%s", (bg,))
            if not cursor.fetchall():
                print(" No donors found")
                return
            confirm = input("Delete all donors with this blood group? (yes/no): ")
            if confirm.lower() == "yes":
                cursor.execute("DELETE FROM donors1 WHERE blood_group=%s", (bg,))
                conn.commit()
                print(" Donors deleted successfully")
            else:
                print(" Delete cancelled")

        elif choice == 3:
            confirm = input("Delete ALL donors? (yes/no): ")
            if confirm.lower() == "yes":
                cursor.execute("DELETE FROM donors1")
                conn.commit()
                print(" All donors deleted")
            else:
                print(" Delete cancelled")

        elif choice == 4:
            print(" Delete cancelled")

        else:
            print(" Invalid choice")

    except Exception as e:
        print("Error:", e)

    finally:
        try:
            cursor.close()
            conn.close()
        except:
            pass


def menu():
    while True:
        print("\n<--- Donor Management System --->")
        print("1. Add Donor")
        print("2. View Donors")
        print("3. Update Donor")
        print("4. Delete Donor")
        print("5. Exit")

        try:
            choice = int(input("Enter your choice: "))
        except:
            print(" Please enter a number")
            continue

        if choice == 1:
            add_student()
        elif choice == 2:
            view_students()
        elif choice == 3:
            update_student()
        elif choice == 4:
            delete_student()
        elif choice == 5:
            print(" Exiting program...")
            break
        else:
            print(" Invalid choice")


menu()         


# HERE ADD,VIEW,UPDATE,DELETE WORKS  HERE UPDATE WORKS  
#WHILE ENTER OPTION 3 I MEAN UPDATE OPTION WHEN CHOOSE THEN  HERE  TABLE DISPLAY...YOU GIVEN NOT GIVEN ID ...HERE SHOWS NOT FOUND....
    #IF YOU GIVEN NUMBER ...CHOOSE OPTIONS  AND ENTER ID THEN EXISTS THEN UPDATE. THEN SUCCESSFULLY COMPLETED UPDATE.
#HERE IN DELETE WHILE CHOOSING OR CALLING DELETE OPTION DISPLAY TABLE THEN CHOOSE OPTIONS ID ,BLOODGROUP,DELETE ALL CANCEL ..WHEN YOU GIVE UNKNOWN NUMBER ID ...IT SHOWS NOT FOUND.
    #NEXT IF YOU GIVEN ID IT SHOWS ID AND ALL OPTIONS THEN DELETED SUCCESSFULLY.
 
```
