```python
#Q6.	Write a  program that accept two integers and display the larger. Also show if the two integers are equal.
def check_large_numbers(num1,num2):
   
    if num1>num2:
        print("num 1 is the larger number")
    elif num2>num1:
        print("num 2 is the larger number")
    else:
        print("the two integers are equal")
num1=int(input("enter the integer number1"))
num2=int(input("enter the integer number2"))        
#Call the function
check_large_numbers(num1,num2)
```

```python
#Q11.	Write a program that takes time as input from user in 24 hours clock format like: 1900 = 7pm. Implement the following case using if, else & else if statements
def check_time(time):
 
    if time>=0000 and time<1200:
        print("good morning")
    elif time>=1200 and time<1700:
        print("good afternoon")
    elif time>=1700 and time<2100:
        print("good evening")
    elif time>=2100 and time<2359:
        print("good night")
    else:
        print("invalid time")
time=int(input("enter the number"))        
#Call the function
check_time(time)
```
