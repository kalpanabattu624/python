```python
#5. Create variables and print their datatypes
def print_datatypes():
    a = 10
    b = 3.14
    c = "Hello"
    d = True
    
    print(a, "datatype:", type(a))
    print(b, "datatype:", type(b))
    print(c, "datatype:", type(c))
    print(d, "datatype:", type(d))

# Call the function
print_datatypes()
```
