```python
'''Q5. Factorial Function
Write a function that computes the factorial of a given number.
Exception Handling Requirements:
•	Handle negative numbers.
•	Handle non-integer inputs.
•	Display appropriate error messages.'''
import math
def factorial_function():
    try:
        number = int(input("enter the number")) 
        if number < 0:
            raise ValueError("Negative number not allowed")
        else:
            factorial=math.factorial(number)
        print(factorial)
    except ValueError:
        print("Enter a valid non-negative integer")   #
    finally:
        print("successfully factorial function completed")      
#call the function
factorial_function()
```

```python
'''Q8. Area of Rectangle
Write a function that calculates the area of a rectangle using:
Pass arguments in two ways:
•	a) As direct values
•	b) As variables
Exception Handling Requirements:
•	Handle zero or negative values.
•	Handle non-numeric inputs.'''
def rectangle_area(length, breadth):
    try:
        if length <= 0 or breadth <= 0:
            raise ValueError("Values must be positive")
        print(length * breadth)
    except ValueError as e:
        print(e)
    finally:
        print("successfully rectangle area function completed")
length = int(input("enter the length"))
breadth = int(input("enter the breadth"))
#call the function
rectangle_area(length, breadth)
```

```python
'''Q9. Power Function
Write a function that computes the power of a number.
Example:
2³ = 8
 Exception Handling Requirements:
•	Handle invalid inputs.
•	Handle large exponent values safely.'''

def power():
    try:
        base = int(input("enter the base"))
        exponent = int(input("enter the exponent"))
        if exponent > 100:
            raise ValueError("Exponent too large")
        print(base ** exponent)
    except ValueError as e:
        print(e)
    finally:
        print("successfully power function completed")

#call the function
power()
```

```python
'''Q10. Sum of List Elements
Write a function that accepts a list of numbers and returns their sum.
Handle:
•	Empty lists
•	Non-numeric values in the list'''

def sum_list_elements():
    try:
        if len(lst) == 0:
            raise ValueError("List is empty")
        total = 0
        for i in lst:
            total += i     
        print(total)
    except ValueError:
        print("List must contain numbers only")
    finally:
        print("successfully list sum function is completed")
lst=[1,3,56,9]
#call the function
sum_list_elements()
```

```python
'''Q7. Hypotenuse Calculation
Write a program to calculate the hypotenuse of a right-angled triangle using functions.
Steps:
1.	Take base and perpendicular as input.
2.	Create a function calculateSquare() to return the square of a number.
3.	Create a function calculateHypotenuse() that uses calculateSquare().
 Exception Handling Requirements:
•	Handle invalid or negative inputs.
•	Handle non-numeric inputs.'''
import math
def calculate_Square(number):
    return number* 2
def calculate_Hypotenuse(base, perpendicular):
    try:
        if base <= 0 or perpendicular <= 0:
            raise ValueError("enter a  positive number")

        hypotenuse = math.sqrt(
            calculate_Square(base) + calculate_Square(perpendicular)
        )
        print(hypotenuse)
    except ValueError:
        print("enter the number must be integer")
    finally:
        print("successfully hypotenuse function completed")
        
base = int(input("enter the base"))
perpendicular = int(input("enter the perpendicular"))
#call the function
# calculate_Square()
calculate_Hypotenuse(base,perpendicular)
```
